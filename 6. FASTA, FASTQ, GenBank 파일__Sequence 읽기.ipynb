# 6. FASTA, FASTQ, GenBank 파일: Sequence 읽기

해당 장에서는 SeqIO.read() 모듈을 활용하여 읽을 예정

SeqIO 모듈은 2가지 메서드를 제공

하나는 SeqIO.parse()

다른 하나는 SeqIO.read()

각 메서드별로 어떤 경우에 사용하는지 알아보고 그 활용법에 대해서 알아볼 예정


## 6.1 SeqIO 모듈로 Sequence 파일을 읽는 두 가지 방법 - FASTA 파일 읽기

### 6.1.1 SeqIO.parse()

SeqIO.parse() 메서드는 인자 2개가 들어감\

첫 번재는 파일 이름, \
두 번째는 첫 번째에 넣어준 인자의 파일 종류

```python
#6.1.1.parse_example_1.py 

from Bio import SeqIO 

seq = SeqIO.parse("sample_1.fasta", "fasta") 
print(type(seq))
for s in seq:
    print(type(s))
    print(s)
    print("") # 줄 구분을 위해 넣었다. 
```

자주 사용하는 파일 포맷

```python
#6.1.1.parse_example_1.py 

from Bio import SeqIO 

seq = SeqIO.parse("sample_2.fasta", "fasta") 
print(type(seq)) 
for s in seq:
    print(type(s))
    print(s)
    print("") # 줄 구분을 위해 넣었다. 
```

### 6.1.2 SeqIO.read() 메서드로 파일 읽기

```python
#6.1.2.read_example_1.py

from Bio import SeqIO 

seq = SeqIO.read("sample_1.fasta", "fasta") 
print(type(seq)) 
print(seq)
```

```python
#6.1.2.read_example_2.py

from Bio import SeqIO 

seq = SeqIO.read("sample_2.fasta", "fasta") 
print(type(seq)) 
print(seq) 
```
